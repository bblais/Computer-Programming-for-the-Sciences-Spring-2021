## Some of the math

$$
\frac{dx}{dt} = \alpha x - \beta xy
$$

$$
\frac{dy}{dt} = \delta xy - \gamma y
$$

If the $\beta$ is zero,

$$
\frac{dx}{dt} = \alpha x
$$


this is the same equation that we've seen for exponential population growth.  if $\delta$ is zero, then 

$$
\frac{dy}{dt} =  - \gamma y
$$

this is the same as exponential population growth with a *negative* growth factor -- population decay.

in a system, when you have a product term like $-\beta x y$ or $\delta x y$, this represents an *interaction* between these two variables.  If $x$ and $y$ are populations, then an interaction is an actual interaction between populations.

- $x$ prey
- $y$ predator

### Fixed points

when the variable is not changing or, in other words, when $dv/dt=0$

### example with logistic growth

$$
\frac{dx}{dt}= ax (1-x/K)
$$



$$
\frac{dx}{dt}=0?
$$

either, $ax=0$ or $(1-x/K)=0$.

this leads immediately, either


$$
x=0 \text{ (unstable)}
$$

$$
x=K \text{ (stable)}
$$


```python
sim=Simulation()
sim.add("x'=a*x*(1-x/K)",29.9,plot=True)
sim.params(a=2,K=30)
sim.run(0,50)
```

## Fixed points for predator prey

$$
\frac{dx}{dt} = \alpha x - \beta xy
$$

$$
\frac{dy}{dt} = \delta xy - \gamma y
$$

$$
\frac{dx}{dt} = 0
$$

$$
\frac{dy}{dt} = 0
$$

$$
\alpha x - \beta xy = 0
$$

$$
\delta xy - \gamma y =0
$$

$$
x \cdot (\alpha  - \beta y) = 0
$$

$$
y \cdot (\delta x - \gamma ) =0
$$

```python
sim=Simulation()
sim.add("x'=*x - *x*y",20,plot=True)
sim.add("y'=*x*y-*y",1,plot=True)
sim.params(=.1,=.1,=.01,=.1)
sim.run(0,500)
```

```python
plot(sim.t,sim.x,label='prey')
plot(sim.t,sim.y,label='predator')
legend()
```

```python
plot(sim.t,sim.x,'b-',label='prey')
legend()
twinx()
plot(sim.t,sim.y,'r-',label='predator')
legend()
```

```python
for  in [0.1,0.15,.2,.25]:
    
    sim=Simulation()
    sim.add("x'=*x - *x*y",20,plot=False)
    sim.add("y'=*x*y-*y",1,plot=False)
    sim.params(=,=.1,=.01,=.1)
    sim.run(0,500)


    plot(sim.t,sim.y,'-',label=f' = {}')
    
legend()
```

the two main effects of changing $\alpha$ are:

1. increasing $\alpha$ increases the maximum predator population
2. increasing $\alpha$ decreases the period of oscillation
