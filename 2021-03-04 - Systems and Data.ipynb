```python
from pyndamics3 import Simulation
```

## Harmonic Oscillator with a non-zero mean

```python
sim=Simulation()
sim.add("x'=v",2,plot=False)
sim.add("v'=-k*x/m",0,plot=False)
sim.params(m=1,k=2)
sim.run(0,50)
```

```python
plot(sim.t,sim.x+5)
```

```python
sim=Simulation()
sim.add("x'=v",2,plot=False)
sim.add("v'=-k*(x-xo)/m",0,plot=False)
sim.params(m=1,k=2,xo=5)
sim.run(0,50)
```

```python
plot(sim.t,sim.x)
```

## Harmonic Oscillator with a non-zero mean which changes over time

```python
sim=Simulation()
sim.add("x'=v",2,plot=True)
sim.add("v'=-k*(x-xo)/m",0,plot=False)
sim.add("xo' = a",5,plot=True)
sim.params(m=1,k=2,a=2)
sim.run(0,50)
```
