```python
from pyndamics3 import Simulation
```

```python
t=array([7,14,21,28,35,42,49,56,63,70,77,84],float)
h=array([17.93,36.36,67.76,98.10,131,169.5,205.5,228.3,247.1,250.5,253.8,254.5])
```

```python
sim=Simulation()
sim.add("y' = a",1,plot=True)
sim.add_data(t=t,y=h,plot=True)  
sim.params(a=5)
sim.run(0,90)
```

```python
from pyndamics3.fit import fit,Parameter
```

```python
results=fit(sim,
   Parameter("a",value=1,min=0))
```

```python
results
```

```python
sim.run(0,90)
```

```python
results=fit(sim,
   Parameter("a",value=1,min=0),
   Parameter("initial_y",value=1,min=0,max=100),
           )
```

```python
results
```

```python
sim.run(0,90)
```

## Logistic

```python
sim=Simulation()
sim.add("y' = a*y*(1-y/K)",1,plot=True)
sim.add_data(t=t,y=h,plot=True)  
sim.params(a=5,K=100)
sim.run(0,90)
```

```python
results=fit(sim,
   Parameter("a",value=1,min=0),
   Parameter("initial_y",value=1,min=0,max=100),
   Parameter("K",value=100,min=0,max=500),
           )
```

```python
results
```

```python
sim.run(0,90)
```

```python
results=fit(sim,
   Parameter("a",value=3,min=0,max=8),
   Parameter("initial_y",value=1,min=0,max=100),
   Parameter("K",value=250,min=0,max=500),
           )

sim.run(0,90)
```

```python
results=fit(sim,
   Parameter("a",value=3,min=0),
   Parameter("initial_y",value=1,min=0,max=100),
   Parameter("K",value=250,min=0,max=500),
           method='powell')

sim.run(0,90)
```

```python
results
```

## Fitting the logistic sample data

```python
import pandas as pd
data=pd.read_csv('data/logistic_sample_data/logistic_sample_data_0.csv')
data.head()
```

```python
t_data=array(data['t'])
y_data=array(data['y'])
```

```python
sim=Simulation()
sim.add("y' = a*y*(1-y/K)",1,plot=True)
sim.add_data(t=t_data,y=y_data,plot=True)  
sim.params(a=5,K=100)
sim.run(0,50)
```

```python
results=fit(sim,
   Parameter("a",value=3,min=0,max=8),
   Parameter("initial_y",value=1,min=0,max=100),
   Parameter("K",value=250,min=0,max=500),
           method='powell')

sim.run(0,90)
```

```python
results
```
