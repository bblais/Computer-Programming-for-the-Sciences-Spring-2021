```python
data=pd.read_csv('data/population_data.txt',sep='\t')
```

```python
#data
```

```python
x=array(data['Year'])
y=array(data['World'])
```

```python
figure(figsize=(6,3))
plot(x,y,'-o')
#xlim([0,2020])
```

```python
sim=Simulation()
sim.figsize=(6,3)  # this line only for when projecting in class for bblais
sim.add("p' = a*p",5,plot=True)
sim.params(a=.003)
sim.run(2020)
```

```python
figure(figsize=(6,3))
plot(x,y,'-o')

plot(sim.t,sim.p,'-')
```

```python
sim=Simulation()
sim.figsize=(6,3)  # this line only for when projecting in class for bblais
sim.add("p' = a*p",5,plot=False)
sim.params(a=.0035)
sim.run(2100)


figure(figsize=(6,3))
plot(x,y,'-o')

plot(sim.t,sim.p,'-')
```

```python
sim=Simulation()
sim.add("p' = a*p",200,plot=False)
sim.params(a=.0015)
sim.run(2100)


plot(x,y,'-o',label='Population Data')

plot(sim.t,sim.p,'-',label='Simple Population Growth Model')
xlabel('Year')
ylabel('Millions of people')
legend()
```

```python
sim=Simulation()
sim.add("p' = a*p",200,plot=False)
sim.add("a' = constant",.0001)
sim.params(constant=1.5e-6)
sim.run(2100)


plot(x,y,'-o',label='Population Data')

plot(sim.t,sim.p,'-',label='Simple Population Growth Model')
xlabel('Year')
ylabel('Millions of people')
legend()
```

```python
plot(sim.t,sim.a)
```

## look at recent data

```python
x=array(data['Year'])
y=array(data['World'])

y=y[x>1950]
x=x[x>1950]

sim=Simulation()
sim.add("p' = a*p",2450,plot=False)
sim.params(a=.018)
sim.run(1950,2020)


plot(x,y,'-o',label='Population Data')

plot(sim.t,sim.p,'-',label='Simple Population Growth Model')
xlabel('Year')
ylabel('Millions of people')
legend()
```

```python
x=array(data['Year'])
y=array(data['World'])

y=y[x>1950]
x=x[x>1950]

sim=Simulation()
sim.add("humans' = growth_rate*humans",2450,plot=False)
sim.params(growth_rate=.018)
sim.run(1950,2020)


plot(x,y,'-o',label='Population Data')

plot(sim.t,sim.humans,'-',label='Simple Population Growth Model')
xlabel('Year')
ylabel('Millions of people')
legend()
```
